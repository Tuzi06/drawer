```python
import cv2
import numpy as np
import pandas as pd
from matplotlib  import pyplot as plt
```

```python
img = cv2.imread('./data/test1.png')
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
mean = np.uint8(np.median(gray))
```

```python
(thresh,baw) = cv2.threshold(gray,mean,255,cv2.THRESH_BINARY)
plt.imshow(baw,cmap='gray')
lower = int(max(0 ,0.7*thresh))
upper = int(min(255,1.3*thresh))
edge = cv2.Canny(baw,lower,upper,apertureSize=7,L2gradient=True)
edge = (edge-255)
plt.imshow(edge,cmap = 'gray')
```

```python
imgr = cv2.bitwise_and(img,img,mask = edge)
plt.imshow(imgr)
cv2.imwrite('fun.png',imgr)
```
